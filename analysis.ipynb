```python
import pandas as pd
import seaborn as sns
from collections import Counter
from matplotlib import pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("data.csv")

# Show the ages with the highest income
plt.figure(figsize=(10, 6))
ax=sns.barplot(data=df, x='Age', y='Total Income', estimator=sum)
plt.title('Total Income by Age')
plt.xlabel('Age')
plt.ylabel('Total Income')
plt.savefig('total_income_by_age.png')

for container in ax.containers:
    ax.bar_label(container,label_type='edge',padding=-1)

# Show the gender distribution across spending categories
expense_categories = ['Utilities', 'Entertainment', 'School Fees', 'Shopping', 'Healthcare']
df_melted = df.melt(id_vars=["Gender"], value_vars=expense_categories, var_name="Expense Category", value_name="Amount")

plt.figure(figsize=(10, 6),dpi=100)
sns.boxplot(data=df_melted, x='Expense Category', y='Amount', hue='Gender',palette='Blues_r')
plt.title('Spending Distribution by Gender')
plt.xlabel('Expense Category')
plt.ylabel('Amount')
plt.xticks(rotation=15)
plt.savefig('spending_distribution_by_gender.png')
```
